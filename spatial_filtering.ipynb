```python
# Averaging filtering

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im = cv.imread('Images/sigiriya.jpg', cv.IMREAD_REDUCED_GRAYSCALE_2) #reduce image size
assert im is not None

k_s = 7
kernel = np.ones((3,3),np.float32)/(k_s*k_s)
imavg = cv.filter2D(im, cv.CV_32F, kernel) #cv_CV32 data type - 32bit floating point numbers

fig, axes = plt.subplots(1,2, sharex='all', sharey='all', figsize=(18,9))
axes[0].imshow(im, cmap='gray')
axes[0].set_title('Original')
axes[0].set_xticks([]), axes[0].set_yticks([])
axes[1].imshow(imavg, cmap='gray')
axes[1].set_title('Averaging')
axes[1].set_xticks([]), axes[1].set_yticks([])
plt.show()


# blury image
```

```python
sobel_x = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
sobel_y = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])

im_x = cv.filter2D(im, cv.CV_64F, sobel_x)
im_y = cv.filter2D(im, cv.CV_64F, sobel_y) #data type is floating point 64bit numbers

fig, ax = plt.subplots(1,2, sharex='all', sharey='all', figsize=(10,10))

# add 125 intensity for change the intensity value of pixels bcz map in range of 0-255
ax[0].imshow(im_x + 125, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Sobel X')

ax[0].set_xticks([]), ax[0].set_yticks([])
ax[1].imshow(im_y + 125, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Sobel Y')
ax[1].set_xticks([]), ax[1].set_yticks([])

plt.show()
```
