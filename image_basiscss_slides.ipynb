```python
#display gray scale pixel

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im = np.zeros((6,8), dtype=np.uint8)
im[2,3] = 255
fig, ax = plt.subplots(1,1,figsize=(6,8))
ax.imshow(im, cmap='gray', vmin=0, vmax=255)
ax.xaxis.set_ticks_position('top')
plt.show()
```

```python
# display color pixel

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im = np.zeros((6,8,3), dtype=np.uint8) #3D array 
im[2,3] = (255, 190, 203)
fig, ax = plt.subplots(1,1,figsize=(6,8))
ax.imshow(im)
ax.xaxis.set_ticks_position('top')
plt.show()
```

```python
im1 = cv.imread('Images/emma.jpg', cv.IMREAD_GRAYSCALE)
assert im1 is not None

im2 = cv.add(im1, 100)
# im2 = im1 + 100  binary values exceed the int 8 bit value

fig, ax = plt.subplots(1, 2, figsize=(6, 8))


ax[0].imshow(im1, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original Image')
ax[1].imshow(im2, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Brightness Increased')
for a in ax:
     a.axis('off')
plt.show()
```

```python
im = cv.imread('Images/emma.jpg')

im_blue = im.copy()
im_blue[:,:,1] = 0
im_blue[:,:,2] = 0

im_red = im.copy()
im_red[:,:,0] = 0
im_red[:,:,1] = 0

im_green = im.copy()
im_green[:,:,0] = 0
im_green[:,:,2] = 0

fig, ax = plt.subplots(2, 2, figsize=(12, 8))
ax[0][0].imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
ax[0][0].set_title('Original Image')

ax[0][1].imshow(cv.cvtColor(im_blue, cv.COLOR_BGR2RGB))
ax[0][1].set_title('Blue Channel')

ax[1][0].imshow(cv.cvtColor(im_red, cv.COLOR_BGR2RGB))
ax[1][0].set_title('Red channel')
ax[1][1].imshow(cv.cvtColor(im_green, cv.COLOR_BGR2RGB))
ax[1][1].set_title('Green Channel')

plt.show()
```
